```python
# KPI Ananalysis
# Total amount of sales
df['TotalAmount'].sum()
```

```python
# minimum total sales
df['TotalAmount'].min()
```

```python
# Sales Analysis
# Total Revenue by state
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=False)
```

```python
# Total sales by category
df.groupby('Category')['TotalAmount'].sum()
```

```python
#Customer and Product analysis
# Top 5 customers based on total
df.groupby('CustomerName')['TotalAmount'].sum().sort_values(ascending=False).head(5)
```

```python
# Total Quantity of product sold
df.groupby('Quantity')['ProductName'].sum()
```

```python
# Payment Analysis
# orders by payment method
df.groupby('PaymentMethod')['OrderID'].count()
```

```python
# Time-Based Analysis
# Orders by date
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df.groupby(df['OrderDate'].dt.date)['OrderID'].count()
```

```python
# Multi-Level Analysis
# sales by state for the category
df.groupby(['State', 'Category'])['TotalAmount'].sum()
```

```python
# Visualization
# Bar Chart showing sales by state
df.groupby('State')['TotalAmount'].sum().plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.title('Sales by State')
plt.show()
```

```python
# Pie Chart showing Category-wise sales
df.groupby('Category')['TotalAmount'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Category-wise Sales')
plt.ylabel('')
plt.show()
```

```python
# Line chart showing monthly sale trend
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
monthly_sales = df.groupby(df['OrderDate'].dt.to_period('M'))['TotalAmount'].sum()
monthly_sales.plot(kind='line')
plt.title('Monthly Sale Trend')
plt.xlabel('Month')
plt.ylabel('Total Amount')
```

```python
# Horizontal Bar chart Top 10 products by sales
df.groupby('ProductName')['TotalAmount'].sum().head(10).plot(kind='bar')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Amount')
```
